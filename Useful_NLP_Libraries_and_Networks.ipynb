 1. How do you perform word tokenization using NLTK and plot a word frequency distribution

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
import matplotlib.pyplot as plt

# Step 1: Download NLTK data if not already done

nltk.download('punkt', download_dir='/root/nltk_data')
nltk.download('punkt_tab', download_dir='/root/nltk_data')


# Sample text
text = "Natural Language Processing is an exciting field of study. NLP is widely used in various applications like text classification, sentiment analysis, and machine translation."

# Step 2: Tokenize the text into words
tokens = word_tokenize(text)

# Step 3: Create a frequency distribution of words
fdist = FreqDist(tokens)

# Step 4: Plot the frequency distribution
plt.figure(figsize=(10, 6))
fdist.plot(30, cumulative=False)  # Plot top 30 most common words
plt.show()

# Optionally, print the frequency distribution
print(fdist)

```

2. How do you use SpaCy for dependency parsing of a sentence ?


```python
import spacy

# Step 1: Load SpaCy's language model (English in this case)
nlp = spacy.load("en_core_web_sm")

# Step 2: Define the sentence to parse
sentence = "SpaCy is an amazing library for NLP."

# Step 3: Parse the sentence using the SpaCy model
doc = nlp(sentence)

# Step 4: Display the dependency parsing results
for token in doc:
    print(f"Word: {token.text}")
    print(f"  - Lemma: {token.lemma_}")
    print(f"  - POS: {token.pos_}")
    print(f"  - Dependency: {token.dep_}")
    print(f"  - Head Word: {token.head.text}")
    print()

# Step 5: Visualize the dependency tree (optional)
from spacy import displacy
displacy.render(doc, style="dep", jupyter=True)

```

3.  How do you use TextBlob for performing text classification based on polarity

```python
from textblob import TextBlob

# Function to classify text based on polarity
def classify_text_based_on_polarity(text):
    # Create a TextBlob object
    blob = TextBlob(text)

    # Get the polarity score
    polarity = blob.sentiment.polarity

    # Classify based on polarity
    if polarity > 0:
        return "Positive"
    elif polarity < 0:
        return "Negative"
    else:
        return "Neutral"

# Test texts
text1 = "I love this product. It's amazing!"
text2 = "This is the worst experience I've had."
text3 = "The movie was okay, not too good or bad."

# Perform classification
print(f"Text 1: {classify_text_based_on_polarity(text1)}")
print(f"Text 2: {classify_text_based_on_polarity(text2)}")
print(f"Text 3: {classify_text_based_on_polarity(text3)}")
```

4.  How do you extract named entities from a text using SpaCy

```python
import spacy

# Load the SpaCy model (ensure you have downloaded it before using)
nlp = spacy.load("en_core_web_sm")

# Text for named entity extraction
text = "Apple Inc. was founded by Steve Jobs and Steve Wozniak in Cupertino, California on April 1, 1976."

# Process the text with SpaCy
doc = nlp(text)

# Extract named entities and print them
for ent in doc.ents:
    print(f"Entity: {ent.text}, Label: {ent.label_}")
```

5.  How can you calculate TF-IDF scores for a given text using Scikit-learn

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Sample corpus of documents (a list of text documents)
documents = [
    "The quick brown fox jumps over the lazy dog.",
    "Never jump over the lazy dog quickly.",
    "The fox is quick and the dog is lazy."
]

# Initialize the TfidfVectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the documents to get the TF-IDF matrix
tfidf_matrix = vectorizer.fit_transform(documents)

# Convert the TF-IDF matrix to a DataFrame for better readability
import pandas as pd
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())

# Display the TF-IDF scores
print(tfidf_df)
```

6.  How do you create a custom text classifier using NLTK's Naive Bayes classifier

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.classify import NaiveBayesClassifier
from nltk.probability import FreqDist

# Download necessary NLTK data files
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# Sample training data (list of tuples containing text and label)
training_data = [
    ("I love this movie", "positive"),
    ("This is a great film", "positive"),
    ("I hated this movie", "negative"),
    ("This film was terrible", "negative"),
    ("I enjoy watching this", "positive"),
    ("It was an awful experience", "negative")
]
```

```python
# Define a function to extract features from a text
def extract_features(text):
    # Tokenize the text into words
    words = word_tokenize(text.lower())

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word.isalpha() and word not in stop_words]

    # Create a set of word features (using word presence as features)
    return {word: True for word in words}
```

```python
# Prepare the training set by extracting features from the text
training_set = [(extract_features(text), label) for text, label in training_data]
```

```python
# Train the Naive Bayes classifier
classifier = NaiveBayesClassifier.train(training_set)
```

```python
# Test the classifier with some sample text
test_data = [
    "I absolutely loved this film",
    "It was such a bad experience",
    "I really enjoyed this movie",
    "This film was boring and bad"
]

# Classify the test data
for text in test_data:
    features = extract_features(text)
    prediction = classifier.classify(features)
    print(f"Text: '{text}' -> Predicted label: {prediction}")
```

9.  How can you create a simple RNN for text classification using Keras

```python
!pip install tensorflow
```

```python
import numpy as np
texts = ['I love programming', 'I hate bugs', 'This is amazing', 'I am frustrated']
labels = np.array([1, 0, 1, 0])  # 1 for positive, 0 for negative
 # 1 for positive, 0 for negative (for example, sentiment classification)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Create a tokenizer
tokenizer = Tokenizer(num_words=10000)  # You can set the number of words you want to consider
tokenizer.fit_on_texts(texts)

# Convert texts to sequences of integers
X = tokenizer.texts_to_sequences(texts)

# Pad sequences to ensure uniform length
X = pad_sequences(X, padding='post')
```

Build the RNN Model


```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense

# Define the model
model = Sequential()

# Embedding layer to map integer sequences to dense vectors
model.add(Embedding(input_dim=10000, output_dim=128, input_length=X.shape[1]))

# Simple RNN layer
model.add(SimpleRNN(units=64))  # 64 RNN units (this can be adjusted)

# Output layer for binary classification (e.g., sentiment: 0 or 1)
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
model.fit(X, labels, epochs=5)
```

10. How do you train a Bidirectional LSTM for text classification

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Bidirectional, LSTM, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.optimizers import Adam

# Example data: texts and corresponding labels (binary classification)
texts = ['I love programming', 'I hate bugs', 'This is amazing', 'I am frustrated']
labels = np.array([1, 0, 1, 0])  # Binary labels: 1 for positive, 0 for negative

# Step 1: Tokenize the text data
tokenizer = Tokenizer(num_words=10000)
tokenizer.fit_on_texts(texts)

# Convert texts to sequences and pad them to a fixed length
X = tokenizer.texts_to_sequences(texts)
X = pad_sequences(X, padding='post')

# Step 2: Define the BiLSTM model
model = Sequential()
model.add(Embedding(input_dim=10000, output_dim=128, input_length=X.shape[1]))
model.add(Bidirectional(LSTM(units=64)))  # Bidirectional LSTM layer
model.add(Dense(1, activation='sigmoid'))  # Output layer for binary classification

# Step 3: Compile the model
model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])

# Step 4: Train the model
model.fit(X, labels, epochs=5, batch_size=2)

# Step 5: Example prediction
new_text = ['I am happy with the results']
new_text_seq = tokenizer.texts_to_sequences(new_text)
new_text_padded = pad_sequences(new_text_seq, padding='post', maxlen=X.shape[1])
prediction = model.predict(new_text_padded)

# Print the predicted class (0 or 1)
print('Predicted class:', prediction)
```

11. How do you implement GRU (Gated Recurrent Unit) for text classification

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, GRU, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.optimizers import Adam

# Example data: texts and corresponding labels (binary classification)
texts = ['I love programming', 'I hate bugs', 'This is amazing', 'I am frustrated']
labels = np.array([1, 0, 1, 0])  # Binary labels: 1 for positive, 0 for negative

# Step 1: Tokenize the text data
tokenizer = Tokenizer(num_words=10000)
tokenizer.fit_on_texts(texts)

# Convert texts to sequences and pad them to a fixed length
X = tokenizer.texts_to_sequences(texts)
X = pad_sequences(X, padding='post')

# Step 2: Define the GRU model
model = Sequential()
model.add(Embedding(input_dim=10000, output_dim=128, input_length=X.shape[1]))  # Embedding layer
model.add(GRU(units=64))  # GRU layer
model.add(Dense(1, activation='sigmoid'))  # Output layer for binary classification

# Step 3: Compile the model
model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])

# Step 4: Train the model
model.fit(X, labels, epochs=5, batch_size=2)

# Step 5: Example prediction
new_text = ['I am happy with the results']
new_text_seq = tokenizer.texts_to_sequences(new_text)
new_text_padded = pad_sequences(new_text_seq, padding='post', maxlen=X.shape[1])
prediction = model.predict(new_text_padded)

# Print the predicted class (0 or 1)
print('Predicted class:', prediction)
```

12.  How do you implement a text generation model using LSTM with Keras

```python
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import LSTM, Dense, Embedding, Dropout
from keras.optimizers import Adam
from keras.utils import to_categorical
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer

# Example corpus
text = '''This is an example corpus of text data. We will use this text to train an LSTM model for text generation.
The model will learn the patterns in the text and generate similar text based on a given seed phrase.'''

# Step 1: Prepare the dataset
# Tokenize the text
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
total_words = len(tokenizer.word_index) + 1  # Vocabulary size

# Convert the text into sequences of word indices
input_sequences = []
for line in text.split('\n'):
    token_list = tokenizer.texts_to_sequences([line])[0]
    for i in range(1, len(token_list)):
        n_gram_sequence = token_list[:i+1]
        input_sequences.append(n_gram_sequence)

# Step 2: Prepare input data (X) and labels (y)
max_sequence_length = max([len(x) for x in input_sequences])  # Length of the longest sequence
input_sequences = pad_sequences(input_sequences, maxlen=max_sequence_length, padding='pre')
X, y = input_sequences[:,:-1], input_sequences[:,-1]

# One-hot encode the labels
y = to_categorical(y, num_classes=total_words)

# Step 3: Build the LSTM model
model = Sequential()
model.add(Embedding(total_words, 100, input_length=X.shape[1]))
model.add(LSTM(150, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(150))
model.add(Dense(total_words, activation='softmax'))

# Step 4: Compile and train the model
model.compile(loss='categorical_crossentropy', optimizer=Adam(), metrics=['accuracy'])
model.fit(X, y, epochs=50, batch_size=64)

# Step 5: Text generation using the trained model
def generate_text(seed_text, model, tokenizer, max_sequence_length, num_words=50):
    output_text = seed_text
    for _ in range(num_words):
        # Tokenize and pad the seed text
        sequence = tokenizer.texts_to_sequences([seed_text])[0]
        sequence = pad_sequences([sequence], maxlen=max_sequence_length, padding='pre')

        # Predict the next word
        predicted_probabilities = model.predict(sequence, verbose=0)
        predicted_word_index = np.argmax(predicted_probabilities)

        # Convert the index back to a word
        predicted_word = tokenizer.index_word[predicted_word_index]

        # Append the predicted word to the output text
        output_text += ' ' + predicted_word

        # Update the seed text for the next prediction
        seed_text += ' ' + predicted_word
    return output_text

# Example usage
seed_text = "This is an"
generated_text = generate_text(seed_text, model, tokenizer, max_sequence_length)
print(generated_text)
```
